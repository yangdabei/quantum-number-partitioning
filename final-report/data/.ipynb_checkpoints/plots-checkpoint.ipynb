```python
%matplotlib inline

# import packages
import numpy as np
import pandas as pd
import seaborn as sns
import csv

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.cm as cm

from cycler import cycler

#------------------------------------------------------------------------------------------------------------
# define plot properties

from matplotlib import rcParams
from matplotlib import rc
from matplotlib.colors import hsv_to_rgb
from mpl_toolkits.axes_grid1 import make_axes_locatable

def rgb(r,g,b):
    return (float(r)/256.,float(g)/256.,float(b)/256.)

cb2 = [rgb(255,127,0), rgb(31,120,180), rgb(51,160,44), rgb(227,26,28), \
       rgb(166,206,227), rgb(253,191,111), rgb(178,223,138), rgb(251,154,153)]

rcParams['figure.figsize'] = (12,12)
rcParams['figure.dpi'] = 250

rcParams['lines.linewidth'] = 1

rcParams['axes.prop_cycle'] = cycler('color', cb2)
rcParams['axes.facecolor'] = 'white'
rcParams['axes.grid'] = False

rcParams['patch.facecolor'] = cb2[0]
rcParams['patch.edgecolor'] = 'white'

rcParams['font.size'] = 30
rcParams['font.weight'] = 200

palette = sns.color_palette("hls", 8)
```

```python
path_n = "results_n.csv"
path_m = "results_m.csv"
```

```python
# fixed n

df_n = pd.read_csv(path_n)

alg = df_n['Algorithm']

fig = plt.figure(figsize=(20,12), dpi=100)
ax1 = fig.add_subplot(111)

boxplots = (alg == "Goemans-Williamson") | (alg == "QAOA") | (alg == "MAQAOA") | (alg == "XQAOA (X=Y)")
sns.boxplot(data=df_n, 
                 x=df_n['m'][boxplots], 
                 y='Cost', 
                 hue=df_n['Algorithm'][boxplots],
                 hue_order=["Goemans-Williamson", "QAOA", "MAQAOA", "XQAOA (X=Y Mixer)"], 
                 palette=[palette[0],palette[1],palette[2],palette[3]])

lineplots = (alg == "Gurobi") | (alg == "Complete Greedy") | (alg == "Complete Karmarkar-Karp")
sns.lineplot(data=df_n,
                  x=df_n['m'][lineplots]-2,
                  y='Cost',
                  hue=df_n['Algorithm'][lineplots],
                  palette=[palette[4],palette[5],palette[6]])

ax2 = ax1.twiny()
ax2.set_xlim(ax1.get_xlim())


fig.subplots_adjust(bottom=0.2)
ax1.set_xlabel(r"$n$")

ax2.set_xlabel(r"$m/n$")

# Move twinned axis ticks and label from top to bottom
ax2.xaxis.set_ticks_position("bottom")
ax2.xaxis.set_label_position("bottom")

# Offset the twin axis below the host
ax2.spines["bottom"].set_position(("axes", -0.075))

# Turn on the frame for the twin axis, but then hide all 
# but the bottom spine
ax2.set_frame_on(True)
ax2.patch.set_visible(False)

for sp in ax2.spines.values():
    sp.set_visible(False)
ax2.spines["bottom"].set_visible(True)


ax2.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax2.xaxis.set_major_formatter(lambda x, pos: str((x+2)/8))



plt.title(f'Cost of Number Partitioning Problems ($n=10$)', fontsize = 18, y=1.0, pad=-45)
plt.yscale('symlog')
plt.xlabel(f'$m$',fontsize = 12)
plt.ylabel('Cost',fontsize = 12)
sns.move_legend(ax, "lower center",
    bbox_to_anchor=(.5,-0.12), ncol=6, title=None, frameon=True)
sns.despine(offset=10, trim=True)

plt.savefig('../report/figures/fixed_n.png', bbox_inches='tight')
```
