```python
# read in CSV from dataset

def readCSV(file):
    df = pd.read_csv(file)#, sep=',', header=None)
    data_columns = [ast.literal_eval(x) for x in list(df.columns)]
    array = df.to_numpy()
    data = []
    for i in range(len(array)):
        data.append([])
        for j in range(len(array[i])):
            data[i].append(ast.literal_eval(array[i][j]))
    
    return np.array(np.moveaxis(data, 0, 1), dtype='object'), data_columns
```

```python
fixed_n, n_array = readCSV("../simulation_dataset/data_fix_n.csv")
fixed_m, m_array = readCSV("../simulation_dataset/data_fix_m.csv")
```

```python
# write in results to CSV

def makeCSV(filepath, var):
    """
    Makes initial file
    """
    with open(filepath, 'w', encoding='UTF8') as f:
        
        writer = csv.writer(f)
        header = [var, 'Cost', 'Algorithm']
        writer.writerow(header)

def writeCSV(filepath, array, alg_str):
    """
    each array contains nested arrays, each representing a particular m, 
    and elements of the nested arrays representing costs 
    
    """
    
    with open(filepath, 'a', encoding='UTF8') as f:
        writer = csv.writer(f)
        if isinstance(array[0],int):
            for i in range(len(array)):
                m = i+2
                cost = array[i]
                writer.writerow([m, cost, alg_str])
        else:
            for i in range(len(array)):
                for j in range(len(array[i])):
                    m = i+2
                    if len(array[i]) != 0:
                        cost = array[i][j]
                        writer.writerow([m, cost, alg_str])
    f.close()   
```
